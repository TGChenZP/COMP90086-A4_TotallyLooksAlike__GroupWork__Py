```python
import matplotlib.pyplot as plt
import numpy as np
```

### semi-hard negatives

```python
train_loss = [0.167, 0.129, 0.049, 0.031, 0.021, 0.020, 0.003, 0.004, 0.003, 0.001, 0.001, 0.002]
test_accuracy = [0.56, 0.543, 0.653, 0.61, 0.647, 0.647, 0.657, 0.673, 0.68, 0.693, 0.693, 0.663]
valid_loss = [0.395, 0.414, 0.386, 0.427, 0.390, 0.395, 0.419, 0.440, 0.393, 0.388, 0.402, 0.411]
```

```python
colour1 = 'slateblue'
colour2 = 'orchid'
SIZE = 14
```

```python
# Create a figure with subplots and shared y-axis
fig, ax1 = plt.subplots()

# Create a twin Axes that shares the same x-axis
ax2 = ax1.twinx()

# Plot test accuracy on the first subplot
ax1.plot(test_accuracy, color=colour1, marker='o', label='Test Accuracy')
ax1.set_xlabel('Epochs', fontsize=SIZE)
ax1.set_ylabel('Test Accuracy', color=colour1, fontsize=SIZE)
ax1.tick_params(axis='y', labelcolor=colour1, labelsize=SIZE)

# Plot train loss on the second subplot
ax2.plot(train_loss, color=colour2, marker='x', label='Train Loss')
ax2.set_ylabel('Train Loss', color=colour2, fontsize=SIZE)
ax2.tick_params(axis='y', labelcolor=colour2, labelsize=SIZE)

ax1.tick_params(axis='x', labelsize=SIZE)
plt.xticks(np.arange(len(train_loss), step=2), range(1, len(train_loss) + 1,2))

ax1.axvline(10, color='red', linestyle='--', label='Max Accuracy', linewidth=1)

# Add a legend for both plots
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='center right', fontsize=SIZE)

# Show the plot
plt.title('Test Accuracy and Train Loss of Semi Hard Negatives', fontsize=SIZE)
plt.show()
```
